```python
# Import libraries
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
from datetime import datetime as dt
import seaborn as sns
```

```python
# Import data
df = pd.read_csv('Data/Prepared Data/NY_merged_agg.csv', index_col = 0)
```

## Setting Theme

```python
# Setting theme for all visualizations
sns.set_theme(style = 'ticks', palette = 'viridis')
```

## Bar Chart - Top 20 Starting Station Frequencies

```python
# Creating a 'value' column to hold the count of frequencies
df['value'] = 1
```

```python
# Grouping by station name, counting the frequency of each, and putting it all in new dataframe
df_t = df.groupby('start_station_name')['value'].count().reset_index()
```

```python
# Checking new dataframe
df_t
```

```python
# Finding top 20 starting stations
top20 = df_t.nlargest(20, 'value')
```

```python
# Checking top20
top20
```

```python
# Plotting top 20 starting stations in a bar chart
sns.barplot(data = top20, x = 'value', y = 'start_station_name')
plt.title("Most Frequent Start Stations")
plt.show()
```

#### The color palette I set up in the theme does not contain enough colors, so it defaukted to this rainbow of colors. In the next plot I will use a "with" statement to alter the color palette for just that plot. 

```python
# Plotting top 20 starting stations in a bar chart with a specific color palette.
plt.figure(figsize = (10, 5))
with sns.color_palette('viridis', n_colors = 20):
    sns.barplot(data = top20, x = 'value', y = 'start_station_name')
    plt.title("Most Frequent Start Stations", fontsize=18)
    plt.ylabel("Start Station Name")
    plt.xlabel("Frequency")
# Saving figure to working folder
plt.savefig('top20_start_stations.png', bbox_inches='tight')
plt.show()
```

## Box Plot

```python
# Creating a box plot of member status compared to daily rides
plt.figure(figsize=(9,5))
sns.boxplot(x = 'member_casual', y = 'daily_rides', data = df, palette = {'member':'lightseagreen', 'casual':'mediumpurple'})
plt.xlabel('Customer Status')
plt.ylabel('Daily Rides')
plt.title('Customer Status vs. Daily Rides')
plt.show()
```

#### This box plot shows the distribution of daily bike rides for each customer type, member and casual. Looking at the median line inside each box, members have a slightly lower average daily ride count, around 3000, compared to casual riders with an average closer to 3500. The interquartile range for members is between approximately 2200 and 3500 while for casual riders it is about 2500 to 3800. There are no outliers for members, but the plot shows some outliers on the lower end of the casual rider box plot indicating a significantly lower than normal ride count. That could be attributed to the fact that casual riders may only rent bikes once. Overall, casual riders have a slightly higher count of daily rides than members, but their central tendency is very similar.   

## FacetGrids

#### A FecetGrid can be used to answer my analysis question: Are more rentals made by casual users or members? And do they typically rent electric bikes or classic bikes? It would make sense to use a FecetGrid because we can compare, side-by-side, the count of members and casual riders. And by plotting the date variable on the x-axis, we can also see how the distribution changes over time, and we can also do the same with the type of bikes that were rented.

```python
# Creating a facetgrid for member status
grid = sns.FacetGrid(df, col = 'member_casual', height=4, aspect=1.5)
grid.map(sns.histplot, 'date', bins = 12)
plt.show()
```

#### In the FacetGrid above we see the comparison between the count of members and casual riders over the year. We can clearly see that there are more members than casual riders throuout the year with the member count being higher in every month of the year. We can also see that number of members and casual riders typically follow the same trend with more rides occurring in the summer months and less during the winter. Another thing I noticed is that members had the highest counts in August and September while casual riders had the highest in July and August. While it makes sense that more bike rentals occur during warmer periods of the year, it's possible that many members could be students that use the bike rental service to get to and from school, which would explain the peak in August and September when school begins and the higher counts into October and November.

```python
# Creating a facetgrid for type of bike rental
grid = sns.FacetGrid(df, col = 'rideable_type', height=4, aspect=1.5)
grid.map(sns.histplot, 'date', bins = 12)
plt.show()
```

#### In the FacetGrid above we are looking at the comparison of the type of bike rented. It is clear to see that people generally prefer to rent classic bikes over electric bikes. This could be due to the increase of health consciousness in today's society where people are more apt to choose an option that is better for their health. Or possibly there are just more classic bikes available than elcetric bikes since electric bike are more expensive to produce and maintain. The trends are fairly similar with the majority of rentals occurring in the summer months, but where the electric bikes peak is in June while the classic bikes peak in August. An idea I had for electric bike rentals being higher in June could be that people on summer vacation, which are predominantly taken in June and July, would rent electric bikes for the fun of it because it's something special to do on vacation or they have never rode on before. Whereas classic bike rentals are a more practical choice and as I mentioned before, a more health conscious choice. There is another chart for docked bikes and there is a small number in the summer months, May - August, but I believe these are bikes that weren't docked properly or that were abandoned and not returned.
